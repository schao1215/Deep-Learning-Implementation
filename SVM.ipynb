```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from time import time
```

```python
X, y = datasets.make_blobs(n_samples=200,
                            n_features = 2,
                            centers = 2, 
                            random_state = 42)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
svc = SVC(kernel="linear")
svc.fit(X, y)
```

```python
X_plot = np.linspace(-6, 6, 100)
y_plot = np.linspace(0, 12, 100)
```

```python
sv = svc.support_vectors_
sv
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(X_plot, (-svc.intercept_[0] - svc.coef_[0, 0]*X_plot) / svc.coef_[0, 1])
plt.scatter(sv[:,0], sv[:,1], c='r')
```

```python
df = pd.read_csv("digits.csv")
y = df.iloc[:, 0]
X = df.iloc[:, 1:]
display(df.head())
display(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression(max_iter=500000, verbose=2)
t = time()
model.fit(X_train, y_train)
print(f"Training time = {time() - t}")
```

```python
model.score(X_test, y_test)
```

```python
model = SVC(max_iter=500000, verbose=2)
t = time()
model.fit(X_train, y_train)
print(f"Training time = {time() - t}")
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```
